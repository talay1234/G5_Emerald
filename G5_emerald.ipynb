# 1. State Problem

### 1.1 Problem and Pain point
ปัญหาการจำแนกงูพิษในประเทศไทยคือ งูพิษ 7 สายพันธุ์ที่พบมากซึ่งประกอบไปด้วย งูสามเหลี่ยม งูแมวเซา งูเขียวหางไหม้ งูจงอาง งูทับสมิงคลา งูกะปะ และงูเห่า งูบางสายพันธุ์มีลักษณะคล้ายกัน ทำให้การระบุด้วยสายตาอาจเกิดความผิดพลาด ซึ่งส่งผลต่อการรักษาและการเลือกใช้เซรุ่มที่ถูกต้อง ดังนั้น กลุ่มของเราจึงพัฒนาโครงการนี้ขึ้นมาโดยมีเป้าหมายเพื่อช่วยแพทย์และเจ้าหน้าที่ สามารถจำแนกงูพิษจากภาพถ่ายได้อย่างแม่นยำและรวดเร็วด้วย AI เพื่อลดความเสี่ยงและเพิ่มประสิทธิภาพในการรักษาผู้ป่วยที่ถูกงูกัด

### 2.2 Data Cleaning
มีการตรวจสอบและทำความสะอาดข้อมูล เพื่อคัดรูปภาพที่ไม่มีคุณภาพ ซึ่งอาจส่งผลเสียต่อการเรียนรู้ของโมเดล

- ตรวจสอบความถูกต้องของ Label: ตรวจสอบรูปภาพในแต่ละคลาส เพื่อไม่ให้รูปภาพจัดประเภทผิดโฟลเดอร์

### 2.3  การแปลงข้อมูล (Data Transformation)
- ปรับขนาดภาพ ให้มีขนาดเท่ากันขนาด  224x224 พิกเซล และทำการปรับค่าสีของแต่ละพิกเซล จากช่วง 0-255 ให้อยู่ในช่วง 0-1 เพื่อให้โมเดลเรียนรู้ได้เร็วและมีเสถียรภาพมากขึ้น

- การเพิ่มข้อมูล (Data Augmentation): เพื่อเพิ่มความหลากหลายให้กับข้อมูล นำมาใช้ โดยการสร้างรูปภาพใหม่ๆ จากรูปภาพเดิมผ่านการแปลงรูปทรง การหมุน (Rotation), การพลิกภาพแนวนอน (Horizontal Flip), การซูม (Zoom), และการเลื่อนตำแหน่งภาพ (Shift) ช่วยลดปัญหา Overfitting

# 3. Train

### 3.1 AI Model: Transfer Learning
- เทคนิคการนำโมเดลที่ถูกฝึกสอนมาอย่างดีแล้วกับชุดข้อมูลขนาดใหญ่ มาเป็นรากฐานหรือจุดเริ่มต้นในการสร้างโมเดลใหม่ โมเดลที่เรียนรู้การจำแนกรูปภาพทั่วไป 1,000 ชนิดจากชุดข้อมูล ImageNet เช่น EfficientNet จะมีความรู้พื้นฐานด้านการมองเห็นอยู่แล้ว

- Base Model (Backbone): คือโมเดล CNN ที่ถูกฝึกมาแล้ว ซึ่งทำหน้าที่ สกัดคุณลักษณะ เรามักจะแช่แข็ง (Freeze) ส่วนนี้ไว้ เพื่อไม่ให้ความรู้พื้นฐานที่เรียนรู้มาอย่างดีแล้วถูกทำลาย

- Classifier Head: คือชั้นเลเยอร์ใหม่ที่เราสร้างขึ้นมาต่อท้าย Base Model เพื่อทำหน้าที่ตัดสินใจและจำแนกประเภทข้อมูลของเราโดยเฉพาะ จากนั้นเราจะฝึกสอน เฉพาะส่วนหัวใหม่นี้ กับข้อมูลของเรา

- ประสิทธิภาพและความเร็ว คือลดระยะเวลาในการฝึกสอน เนื่องจากโมเดลมีความรู้พื้นฐานมาแล้ว จึงใช้เวลาในการเรียนรู้โจทย์ใหม่น้อยลงอย่างมาก

- การปรับจูน (Fine-Tuning) หลังจากฝึกสอนส่วนหัวใหม่แล้ว อาจมีการ Unfreeze บางส่วนของ Base Model แล้วทำการฝึกสอนโมเดลทั้งหมดอีกครั้งด้วยอัตราการเรียนรู้ (Learning Rate) ที่ต่ำมากๆ เพื่อให้โมเดลพื้นฐานสามารถปรับตัวเข้ากับลักษณะเฉพาะของข้อมูลชุดใหม่ของเราได้ดียิ่งขึ้น

### 3.2 Hyperparameters
- np.random.seed(42) และ tf.random.set_seed(42) กำหนดจุดเริ่มต้น ของการสุ่มตัวเลขในไลบรารี NumPy และ TensorFlow ทำให้ผลการทดลองสามารถทำซ้ำได้ให้เป็นค่าเดียวกันทุกครั้งที่รันโค้ด
- IMAGE_SIZE = 224 กำหนดขนาดของรูปภาพ (ความกว้าง x ความสูง) ที่จะป้อนเข้าโมเดล
- BATCH_SIZE = 8 กำหนดจำนวนรูปภาพที่จะใช้ในการฝึกโมเดลในแต่ละรอบย่อย โมเดลจะดูรูปภาพ 8 รูป, คำนวณค่าความผิดพลาดเฉลี่ยของทั้ง 8 รูป, แล้วจึงอัปเดตค่าน้ำหนัก (weights) หนึ่งครั้ง
- EPOCHS = 50 โมเดลเรียนรู้ข้อมูลซ้ำไปซ้ำมาทั้งหมด 50 รอบ เพื่อให้โมเดลค่อยๆ ปรับปรุงความแม่นยำและลดค่าความผิดพลาดลงเรื่อยๆ
- adam_opt = Adam(learning_rate=1e-4) ค่า 1e-4 (เท่ากับ 0.0001) เป็น learning rate ที่ค่อนข้างต่ำ เหมาะสำหรับการปรับจูนโมเดล (Fine-tuning)  เพื่อไม่ให้ความรู้เดิมที่โมเดลมีอยู่แล้วเสียหายไป


```python
np.random.seed(42)
tf.random.set_seed(42)
IMAGE_SIZE = 224
BATCH_SIZE = 8
EPOCHS = 50
adam_opt = Adam(learning_rate=1e-4)
```

### 3.3 Model Training
- อัตราส่วนข้อมูล train : test = 70 : 30

```python
#Rescale ข้อมูล Test
test_datagen = ImageDataGenerator(rescale=1./255)

#Rescale และทำ Data Augmentation ข้อมูล Train
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest',
    validation_split=0.15
)

#สร้าง Generator เพื่อดึงข้อมูลจากโฟลเดอร์
train_generator = train_datagen.flow_from_directory(
    train_folder,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_folder,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    color_mode='rgb',
    shuffle=False
)
```

### โหลด Base Model EfficientNetV2S

```python
base_model = tf.keras.applications.EfficientNetV2S(
    weights='imagenet',
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3),
    include_top=False
)
```

### สร้างโมเดล ด้วยเทคนิค Transfer Learning

```python
# Freeze base model
base_model.trainable = False

model = tf.keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),

    layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.5),

    layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.4),

    # Output Layer
    layers.Dense(num_classes, activation='softmax')
])
```

### กำหนดค่า Compile สำหรับโมเดล
- Optimizer Adam ปรับ learning rate ของแต่ละพารามิเตอร์
- categorical_crossentropy จำแนกประเภทข้อมูลแบบหลายคลาส (multi-class classification) โดยเฉพาะ และมีข้อกำหนดว่า ข้อมูล label (คำตอบจริง) จะต้องอยู่ในรูปแบบ One-Hot Encoding
- Metrics แสดงผลเมตริกความแม่นยำ (accuracy) หลังจากแต่ละ epoch

```python
model.compile(
    optimizer=Adam(1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

### เทรนโมเดลรอบแรก (Warm-up)

```python
history_warm = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=8,
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
    ]
)
```

### Fine-tuning

```python
# warm up freeze
base_model.trainable = True

# compile the model
model.compile(
    optimizer=adam_opt,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

# 5. Model Improvement

### 5.1 การวิเคราะห์ข้อบกพร่องของโมเดล
**Data Augmentation**:
- การเพิ่มข้อมูล ยังไม่เพียงพอที่จะทำให้โมเดลทนทาน ต่อความหลากหลายของข้อมูล

**Model Architecture** :
- ขาด Batch Normalization ทำให้การฝึกสอนไม่เสถียร
- Dropout Rate ที่สูงเกินไป ทำให้โมเดลเรียนรู้ได้ไม่เต็มที่

**Training** :
- Learning Rate ในช่วง Warm-up ต่ำไป ทำให้ต้องเรียนรู้จากศูนย์
- Patience ต่ำเกินไป ทำให้การฝึกสอนสิ้นสุดเร็วเกินไป
- Monitor ที่ val_loss อย่างเดียว

### 5.2 การปรับปรุงโมเดล
**Data Augmentation** :
- rotation_range เพิ่มจาก 20 เป็น 30
- width_shift_range และ height_shift_range เพิ่มจาก 0.2 เป็น 0.3
- zoom_range เพิ่มจาก 0.2 เป็น 0.3
- brightness_range เปลี่ยนจาก [0.8, 1.2] เป็น [0.7, 1.3]

**Model Architecture** :
- เพิ่ม layers.BatchNormalization() เข้ามาหลัง GlobalAveragePooling2D
- ลด Dropout จาก 0.5 และ 0.4 เหลือ 0.3 และ 0.2

**Training** :
- ปรับ Learning Rate ช่วง Warm-up ใช้ Adam(learning_rate=1e-3)
- เพิ่ม patience=5 และ EarlyStopping เพิ่มเป็น patience=10
- เปลี่ยนเกณฑ์การบันทึกโมเดล: ModelCheckpoint เปลี่ยนเป็น monitor="val_accuracy" และตั้ง mode='max'

### เทรนโมเดลทีได้รับการปรับปรุง

```python
BATCH_SIZE = 8
IMAGE_SIZE = 224
VAL_SPLIT = 0.15
WARMUP_EPOCHS = 8
FINE_TUNE_EPOCHS = 50
BEST_MODEL_FILE = './final_model1.h5'
BEST_MODEL_FILE_IMPROVED = './final_model1-improved.h5'
```

### DATA AUGMENTATION

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,        # Increased from 20
    width_shift_range=0.3,    # Increased from 0.2
    height_shift_range=0.3,   # Increased from 0.2
    shear_range=0.2,
    zoom_range=0.3,           # Increased from 0.2
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.7, 1.3],  # Wider range
    fill_mode='nearest',
)

test_datagen = ImageDataGenerator(
    rescale=1./255,
)
```

### DATA GENERATORS

```python
train_generator = train_datagen.flow_from_directory(
    train_folder,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_folder,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
```

### MODEL BUILDING

```python
# Load pre-trained EfficientNetV2S
base_model = tf.keras.applications.EfficientNetV2S(
    include_top=False,
    weights='imagenet',
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)
)

# Freeze base model for warmup
base_model.trainable = False

model = tf.keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),  # Added for stability

    layers.Dense(256, activation='relu',
                 kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.3),  # Reduced from 0.5

    layers.Dense(128, activation='relu',
                 kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    layers.Dropout(0.2),  # Reduced from 0.4

    # Output layer
    layers.Dense(num_classes, activation='softmax')
])

# Compile for warmup phase
model.compile(
    optimizer=Adam(learning_rate=1e-3),  # Higher LR for warmup 3
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

### PHASE 1: WARMUP TRAINING

```python
print("\n" + "="*50)
print("PHASE 1: WARMUP TRAINING (Base Model Frozen)")
print("="*50 + "\n")

history_warmup = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=WARMUP_EPOCHS,
    callbacks=[
        EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True,
            verbose=1
        )
    ],
    verbose=1
)
```

### PHASE 2: FINE-TUNING & เก็บโมเดลที่ได้รับการปรับปรุงใหม่ไว้ในไฟล์ชื่อ final_model1_improved.h5

```python
# warm up freeze
base_model.trainable = True
adam_opt = Adam(learning_rate=1e-4)

# compile the model
model.compile(
    optimizer=adam_opt,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
callbacks = [
    ModelCheckpoint(
        BEST_MODEL_FILE_IMPROVED,
        verbose=1,
        save_best_only=True,
        monitor="val_accuracy",  # Monitor accuracy instead of loss
        mode='max',
        save_weights_only=False
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        patience=5,      # Increased patience
        factor=0.3,      # Less aggressive reduction
        verbose=1,
        min_lr=1e-7,
    ),
    EarlyStopping(
        monitor='val_loss',
        patience=10,     # Increased patience
        restore_best_weights=True,
        verbose=1
    )
]

# Train the model
history_finetune = model.fit(
    train_generator,
    epochs=FINE_TUNE_EPOCHS,
    validation_data=test_generator,
    callbacks=callbacks,
    verbose=1
)
```

### SAVE FINAL MODEL

```python
print("\n" + "="*50)
print("Training Complete!")
print("="*50)
print(f"Best model saved to: {BEST_MODEL_FILE_IMPROVED}")

# Print final metrics
final_train_acc = history_finetune.history['accuracy'][-1]
final_val_acc = history_finetune.history['val_accuracy'][-1]
final_train_loss = history_finetune.history['loss'][-1]
final_val_loss = history_finetune.history['val_loss'][-1]

print(f"\nFinal Training Accuracy: {final_train_acc:.4f}")
print(f"Final Validation Accuracy: {final_val_acc:.4f}")
print(f"Final Training Loss: {final_train_loss:.4f}")
print(f"Final Validation Loss: {final_val_loss:.4f}")
```

### กราฟ loss rate และ accuracy rate ของโมเดลทีไ่ด้รับการปรับปรุง

```python
def plot_training_history(warmup_history, finetune_history):

    # Combine histories
    warmup_epochs = len(warmup_history.history['accuracy'])

    train_acc = warmup_history.history['accuracy'] + finetune_history.history['accuracy']
    val_acc = warmup_history.history['val_accuracy'] + finetune_history.history['val_accuracy']
    train_loss = warmup_history.history['loss'] + finetune_history.history['loss']
    val_loss = warmup_history.history['val_loss'] + finetune_history.history['val_loss']

    epochs_range = range(len(train_acc))

    plt.figure(figsize=(14, 5))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, train_acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.axvline(x=warmup_epochs-1, color='r', linestyle='--', label='Fine-tuning Start')
    plt.legend(loc='lower right')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.grid(True, alpha=0.3)

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, train_loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.axvline(x=warmup_epochs-1, color='r', linestyle='--', label='Fine-tuning Start')
    plt.legend(loc='upper right')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('training_history_improved.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("\nTraining history plot saved as 'training_history_improved.png'")

# Plot the results
plot_training_history(history_warmup, history_finetune)
```
